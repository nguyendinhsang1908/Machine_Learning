```
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
import numpy as np
import select as sl
```

```
A = array([[1,2], [3,4], [5,6]])
print(A)
```

```
M = mean(A.T, axis=1) ##giá trị trung bình của từng cột ma trận
print(M)
```

```
C = A - M # căn giữa(so sánh) các giá trị từng cột với giá trị trung bình
print(C)
```

```
V = cov(C.T) ## phương sai ma trận
print(V)
```

```
values, vectors = eig(V) # tính các vector riêng và giá trị riêng
```

```
print(vectors)
```

```
print(values) #từ giá trị này ta thấy vector thứ 2 hầu như không đóng góp quá nhiều dữ liệu nên ta có thể lược bỏ 
```

```
P = vectors.T.dot(C.T)
print(P.T)
```

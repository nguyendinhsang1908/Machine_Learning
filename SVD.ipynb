```
from numpy import array
from numpy import diag
from numpy import dot
from numpy import zeros
from scipy.linalg import svd
```

```
A = array([[1,1], [1, 1]])
print(A)
```

```
U, s, VT = svd(A) ##sử dụng hàm 
print(U)
```

```
print(s)
```

```
print(VT)
```

```
Sigma = zeros((A.shape[0], A.shape[1]))## tạo ma trận với kích thước bằng ma trận A
```

```
print(Sigma)
```

lấy các giá trị của s cho vào đường chéo 

```
Sigma[:A.shape[1], :A.shape[1]] = diag(s) ##lấy các giá trị từ s gán cho sigma
```

```
print(Sigma)
```

Thử lại





```
B = U.dot(Sigma.dot(VT))
```

```
print(B)
```

ma trận giả nghịch đảo


```
for i in range (0,s.shape[0]):
  if s[i]!=0:
    s[i]=1/s[i];
```

```
print(s)
```

```
Sigma = zeros((A.shape[0], A.shape[1]))
```

```
Sigma[:A.shape[1], :A.shape[1]] = diag(s) 
```

```
print(Sigma)
```

```
B = (VT.T).dot(Sigma.dot(U.T)) ##giả nghịch đảo ma trận a
print(B)
```
